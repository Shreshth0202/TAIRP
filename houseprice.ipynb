```python
cat_func = ['city']
```

```python
catdata = df[cat_func]
```

```python
changedcat = pd.get_dummies(catdata , columns=cat_func , drop_first=True)
```

```python
df = pd.concat([df,changedcat], axis=1)
```

```python
df.drop(columns = cat_func , inplace=True)
```

```python
df.to_csv("engineered_data" , index=False)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df.drop(columns=['date','year'], inplace=True)
```

```python
df.dtypes
```

```python
df.drop(columns=['street','statezip','country'], inplace=True)
```

```python
if 'city' in df:
    df['street','statezip'] = df['city'].str.split('',1,expand=True)
    df.drop(columns='country',inplace=True)
else:
    print("The City column doesn't Exist in Data set!!")
```

Modelling


```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['price'])
y = df['price']
yval = np.log(df['price'])
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
pred = lr.predict(x_test)
```
