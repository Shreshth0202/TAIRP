Data Cleaning

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['vote_count'] = pd.to_numeric(df['vote_count'], errors='coerce')
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
```

```python
# Assuming df is your DataFrame

# Filtering based on conditions
df = df[(df['budget'] >= 6500000) & (df['budget'] <= 45000000) &
        (df['popularity'] >= 4) & (df['popularity'] <= 25) &
        (df['revenue'] >= 10000000) & (df['revenue'] <= 90000000) &
        (df['vote_count'] >= 50) & (df['vote_count'] <= 700) &
        (df['vote_average'] >= 1)]

# Resetting index after dropping rows
df.reset_index(drop=True, inplace=True)
```

Data Vizualization

```python
sn.histplot(df.runtime , color="Pink" , bins=20)
```

```python
plt.figure(figsize=(10,6))
sn.countplot(x="original_language" , data= df , order=df['original_language'].value_counts().index)
plt.title("No. of Movies W.R.T to Language")
plt.xlabel("Language")
plt.ylabel("Count")
plt.xticks(rotation = 90)
plt.show
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df.sort_values(by='release_date' , inplace = True)
plt.plot(df['release_date'],df['budget'],marker = 'o')
plt.xlabel("Release date")
plt.ylabel("Amount(Millions)")
plt.show()
```

```python
sn.scatterplot(x='budget' , y='revenue', data = df )
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.title("Budget VS Revenue")
plt.show()
```

```python
numeric_columns = ['budget', 'popularity', 'revenue', 'vote_count', 'vote_average']
corrmat= df[numeric_columns].corr()
sn.heatmap(corrmat , vmax=.9, square=True,annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show
```

```python
sn.boxplot(x='original_language', y='budget',data=df)
plt.title("Language VS Budget")
plt.xlabel("Language")
plt.ylabel("Budget(Millions)")
plt.xticks(rotation = 90)
plt.show
```
